```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# CPU simple sampling Processing time with different N
x = [64, 128, 256, 512, 1024, 2048, 4096, 8192, 16384, 32768, 65536]
process_time = [0.001344, 0.001344, 0.001792, 0.003072, 0.005472, 0.009600, 0.019072, 0.037440, 0.074048, 0.147264, 0.295776]
plt.plot(x, process_time)
plt.xscale('log')
plt.xlabel('Sample point amount(log)')
plt.ylabel('Elapsed time(ms)')
plt.title('CPU simple sampling Processing time with different N')
plt.savefig('cpu_ss.png', dpi=300)
plt.show()
```
