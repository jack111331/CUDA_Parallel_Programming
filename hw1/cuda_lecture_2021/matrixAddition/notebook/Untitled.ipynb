```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [4, 8, 10, 16, 20, 32]
input_time = [52.341919, 52.140350, 52.190689, 52.234783, 52.151104, 52.183552]
plt.plot(x, input_time)
plt.xlabel('Block size per dimension')
plt.ylabel('Elapsed time(ms)')
plt.title('GPU Input time')
plt.savefig('gpu_input_time.png', dpi=300)
plt.show()
```
