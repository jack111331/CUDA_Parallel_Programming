```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1 GPU vector dot product fixing grid size = 64
x = [2, 4, 8, 16, 32, 64, 128, 256, 512]
processing_time = [169.934753, 92.518082, 49.670368, 25.108959, 12.613216, 6.421568, 3.872736, 2.761472, 2.840352]
plt.plot(x, processing_time)
plt.xscale('log')
plt.xlabel('Block size(log)')
plt.ylabel('Elapsed time(ms)')
plt.title('1 GPU Processing time fixing grid size=64')
plt.savefig('1gpu_processing_time_fix_grid.png', dpi=300)
plt.show()
```
