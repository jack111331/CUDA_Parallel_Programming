```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# GPU without Texture memory
x = [2, 4, 8, 16, 32]
total_time = [163276.171875, 53498.398438, 20418.162109, 17945.771484, 23617.394531]
plt.plot(x, total_time)
plt.xlabel('Block size')
plt.ylabel('Elapsed time(ms)')
plt.title('GPU Total time without texture memory')
plt.savefig('gpu_total_time.png', dpi=300)
plt.show()
```

```python
# GPU with Texture memory
x = [2, 4, 8, 16, 32]
total_time = [134234.281250, 43675.523438, 20931.169922, 18244.033203, 24534.421875]
plt.plot(x, total_time)
plt.xlabel('Grid size')
plt.ylabel('Elapsed time(ms)')
plt.title('GPU Total time with texture memory')
plt.savefig('gpu_total_time_texture.png', dpi=300)
plt.show()
```
