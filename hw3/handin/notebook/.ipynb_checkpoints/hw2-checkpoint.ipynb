```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# GPU without Texture memory
x = [2, 4, 8, 16, 32]
total_time = [398.956482, 230.442108, 147.089493, 101.582596, 78.679008, 67.385117, 68.799171, 58.835869, 58.930176]
plt.plot(x, total_time)
plt.xlabel('Block size')
plt.xscale('log')
plt.ylabel('Elapsed time(ms)')
plt.title('GPU Total time without texture memory, x axis in log scale')
plt.savefig('gpu_total_time.png', dpi=300)
plt.show()
```

```python
# GPU with Texture memory
x = [2, 4, 8, 16, 32]
total_time = [398.832764, 231.020996, 146.677505, 101.716064, 78.679008, 68.067039, 68.335007, 59.072803, 60.990364]
plt.plot(x, total_time)
plt.xlabel('Grid size')
plt.xscale("log")
plt.ylabel('Elapsed time(ms)')
plt.title('GPU Total time with texture memory, x axis in log scale')
plt.savefig('gpu_total_time_texture.png', dpi=300)
plt.show()
```
