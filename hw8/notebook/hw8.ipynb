```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# 1 GPU Monte Carlo simulation of 2D Ising model on the torus using different block size
x = [1, 2, 5, 10]
process_time = [19503.261719, 18496.666016, 17410.759766, 17176.330078]
plt.plot(x, process_time)
plt.xlabel('Block size')
plt.ylabel('Processing time(ms)')
plt.title('Processing time of 1 GPU Monte carlo simulation of 2D Ising model\nwith different block size')
plt.savefig('1gpu_block.png', dpi=300)
plt.show()
```

```python
# 2 GPU Monte Carlo simulation of 2D Ising model on the torus using different block size
x = [1, 5]
process_time = [17760.126953, 17267.140625]
plt.plot(x, process_time)
plt.xlabel('Block size')
plt.ylabel('Processing time(ms)')
plt.title('Processing time of 2 GPU Monte carlo simulation of 2D Ising model\nwith different block size')
plt.savefig('2gpu_block.png', dpi=300)
plt.show()
```

```python
# 1 GPU T=2.0
T_list = ["2.0", "2.1", "2.2", "2.3", "2.4", "2.5"]
for T in T_list:
    df = pd.read_csv('../handin/Ising2D_NGPU/result/GPU_1/T_'+T+'/Output_1.txt') 
    plt.plot(df['Sweep'], df['Energy density'])
    plt.xlabel('sweep')
    plt.ylabel('<E>')
    plt.title('Energy density of 1 GPU Monte carlo simulation of 2D Ising model\nwith temperature is %s' % T)
    plt.savefig('1gpu_%s_E.png' % T, dpi=300)
    plt.show()
    
    plt.plot(df['Sweep'], df['Magnetization density'])
    plt.xlabel('sweep')
    plt.ylabel('<M>')
    plt.title('Magnetization density of 1 GPU Monte carlo simulation of 2D Ising model\nwith temperature is %s' % T)
    plt.savefig('1gpu_%s_M.png' % T, dpi=300)
    plt.show()
```

```python
# 2 GPU
T_list = ["2.0", "2.1", "2.2", "2.3", "2.4", "2.5"]
for T in T_list:
    df = pd.read_csv('../handin/Ising2D_NGPU/result/GPU_2/T_'+T+'/Output_1.txt') 
    plt.plot(df['Sweep'], df['Energy density'])
    plt.xlabel('sweep')
    plt.ylabel('<E>')
    plt.title('Energy density of 2 GPU Monte carlo simulation of 2D Ising model\nwith temperature is %s' % T)
    plt.savefig('2gpu_%s_E.png' % T, dpi=300)
    plt.show()
    
    plt.plot(df['Sweep'], df['Magnetization density'])
    plt.xlabel('sweep')
    plt.ylabel('<M>')
    plt.title('Magnetization density of 2 GPU Monte carlo simulation of 2D Ising model\nwith temperature is %s' % T)
    plt.savefig('2gpu_%s_M.png' % T, dpi=300)
    plt.show()
```

```python
# 2 GPU Absolute magnetization over different temperature
x = [2.0, 2.1, 2.2, 2.3, 2.4, 2.5]
absolute_magnetization = [9.115414e-001, 8.686338e-001, 7.833565e-001, 1.374410e-001, 6.033856e-003, 3.581217e-003]
plt.plot(x, absolute_magnetization)
plt.xlabel('Temperature')
plt.ylabel('Absolute magnetization')
plt.title('Absolute magnetization with different temperature')
plt.savefig('2gpu_absolute_magnetization.png', dpi=300)
plt.show()
```
