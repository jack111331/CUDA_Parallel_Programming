```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# GPU using global memory generating exponential distribution, fixing grid size = 32
x = [2, 4, 8, 16, 32, 64, 128, 256, 512]
total_time = [543.889221, 298.496216, 180.342194, 118.567711, 108.088638, 102.202499, 101.012672, 102.990242, 101.273087]
plt.plot(x, total_time)
plt.xscale('log')
plt.xlabel('Block size(log)')
plt.ylabel('Elapsed time(ms)')
plt.title('GPU using global memory Total time, fixing grid size=32')
plt.savefig('gpu_gmem_total_time_fix_grid.png', dpi=300)
plt.show()
```

```python
# GPU using shared memory generating exponential distribution, fixing grid size = 32
x = [2, 4, 8, 16, 32, 64, 128, 256, 512]
total_time = [500.174225, 297.700714, 179.366241, 118.832611, 89.845215, 73.581085, 64.942818, 60.893345, 59.796993]
plt.plot(x, total_time)
plt.xscale('log')
plt.xlabel('Block size(log)')
plt.ylabel('Elapsed time(ms)')
plt.title('GPU using shared memory Total time, fixing grid size=32')
plt.savefig('gpu_shmem_total_time_fix_grid.png', dpi=300)
plt.show()
```

```python
x = np.arange(0.5, 19.5, 1.0)
y = np.array([51781403, 19051914, 7009657, 2577159, 948336, 348861, 127730, 47311, 17448, 6347, 2460, 873, 316, 115, 45, 12, 9, 4, 0])
y = np.multiply(y, 1.0 / y.sum())
plt.plot(x, y, label='Random sample in exponential distribution')
x = np.arange(0.0, 19.5, 0.1)
y = np.exp(-x)
plt.plot(x, y, label='Real exponential distribution')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Histogram showing exponential distribution')
plt.legend()
plt.savefig('histogram.png', dpi=300)
plt.show()
```
