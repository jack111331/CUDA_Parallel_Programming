```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [2, 4, 8, 16, 32, 64, 128, 256, 512]
total_time = [398.956482, 230.442108, 147.089493, 101.582596, 78.679008, 67.385117, 68.799171, 58.835869, 58.930176]
plt.plot(x, total_time)
plt.xlabel('Block size')
plt.xscale('log')
plt.ylabel('Elapsed time(ms)')
plt.title('GPU Total time fixing grid size=32 with x axis in log scale')
plt.savefig('gpu_total_time_fixing_grid_size.png', dpi=300)
plt.show()
```
